1. Зчитані дані - це сума покупок в доларах наших покупців на сайті протягом липня 2022 року.

До цього місяця, ми мали продажі в середньому на чек $20. Чи вийшло у нас статистично значущо **підвищити** середній чек за останній місяць?

Оберіть статистичний тест (z-test чи t-test). Виведіть p-значення статистичного тесту та зробіть висновок, чи ми маємо зараз статистично значущу різницю у розмірі середнього чеку.

**Підказка**: тут тест буде одновибірковий.

```python
from statsmodels.stats.weightstats import ztest
#z-test бо даних більше 30 
ztest_Score, p_value = ztest(data, value = 20, alternative='larger')
ztest_Score, p_value
```

Використано код для одновибіркового z-тесту, де ми перевіряємо гіпотезу:

Нульова гіпотеза H0 середній чек дорівнює 20 доларів
Альтернативна гіпотеза На середній чек статистично значущо більший за 20 доларів

Z-статистика: -0.9953477389335053
Це значення z-тесту показує відхилення середнього значення зразка від гіпотетичного значення $20 в одиницях стандартних відхилень.

p-значення: 0.8402164165464012
Це значення показує ймовірність того, що ми спостерігаємо отриманий або ще більш екстремальний результат, якщо нульова гіпотеза правдива.
Оскільки p-значення 0.84 значно більше за звичайний поріг (наприклад, 0.05 або 0.01), ми не можемо відкинути нульову гіпотезу.

Висновок:
На основі результатів z-тесту, з p-значенням 0.84, ми не маємо підстав стверджувати, що середній чек за липень 2022 року статистично значущо перевищує $20. Тобто, статистичних доказів підвищення середнього чеку немає.

  **Підказка**: треба провести двовибірковий ztest з Н0 - що середні значення в липні та серпні 2022 року - однакові. Для кращого розуміння ситуації, можливо вам буде цікаво переглянути, які середні показники чеку були в кожен з місяців. Порівняйте свій висновок з простого спостереження середніх значень і висновок статистичного тесту. Чи вони збігаються?

```python
df_2 = pd.read_csv('data/data_2.csv')
```

```python
ztest_score, p_value = ztest(df, df_2, alternative='two-sided')

print("Середній чек липня:", df.mean())
print("Середній чек серпня:", df_2.mean())
print(f"Z-тест: {ztest_score}, p-значення: {p_value}")
```



Висновок:

За p-значенням 0.9958, ми не маємо підстав відкидати нульову гіпотезу. Тобто, немає статистично значущої різниці між середніми чеками за липень і серпень 2022 року.

Збіг висновків: Середні значення за липень і серпень дуже близькі (19.38 проти 19.37 доларів). Це узгоджується з результатами статистичного тесту — немає суттєвої різниці між середніми значеннями.
