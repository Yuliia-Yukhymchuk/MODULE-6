# Завдання 1


Щоб знайти ймовірність того, що людина насправді хвора, якщо тест показав позитивний результат, використаємо формулу Баєса:

$$
P(H | T^+) = \frac{P(T^+ | H) \cdot P(H)}{P(T^+)}
$$

Де:
$$ 
P(H)
$$ — ймовірність того, що людина хвора (глобальна ймовірність хвороби) = 1% = 0.01 
$$P(\neg H)
$$— ймовірність того, що людина здорова = 99% = 0.99
$$P(T^+ | H)
$$— ймовірність позитивного тесту за умови, що людина хвора (чутливість тесту) = 99% = 0.99
$$P(T^+ | \neg H)
$$— ймовірність позитивного тесту за умови, що людина здорова (хибнопозитивний результат) = 5% = 0.05

Знайдемо ймовірність позитивного тесту \( P(T^+) \) за формулою повної ймовірності:

$$
P(T^+) = P(T^+ | H) \cdot P(H) + P(T^+ | \neg H) \cdot P(\neg H)
$$

Підставимо значення:

$$
P(T^+) = (0.99 \cdot 0.01) + (0.05 \cdot 0.99)
$$

Обчислимо \( P(T^+) \):

$$
P(T^+) = 0.0099 + 0.0495 = 0.0594
$$

Тепер підставимо все у формулу Баєса:

$$
P(H | T^+) = \frac{0.99 \cdot 0.01}{0.0594}
$$

Обчислимо результат:

$$
P(H | T^+) \approx 0.1667
$$

Отже, ймовірність того, що людина насправді хвора, якщо тест показав позитивний результат, становить приблизно **16.67%**.


# Завдання 2

Позначимо:
- $P(C) = 0.7$ — ймовірність чесності
- $P(L) = 0.3$ — ймовірність брехні
- $P(T^+ | C) = 0.1$ — ймовірність позитивного результату (хибнопозитивний)
- $P(T^+ | L) = 1$ — ймовірність позитивного результату для брехунів

### Формули
1. Знайдемо $P(T^+)$:

$$
P(T^+) = (1 \cdot 0.3) + (0.1 \cdot 0.7) = 0.3 + 0.07 = 0.37
$$

2. Використаємо формулу Баєса:

$$
P(L | T^+) = \frac{1 \cdot 0.3}{0.37} \approx 0.8108
$$

# Завдання 5


### Позначення
- $A$ — подія, що на вхід подано цифру "3".
- $B$ — подія, що програма не розпізнала цифру "3".

### Задані ймовірності
- $P(A) = 0.2$ — ймовірність, що на вхід подано цифру "3".
- $P(A') = 1 - P(A) = 0.8$ — ймовірність, що на вхід подано не цифру "3".
- $P(B | A) = 1 - 0.95 = 0.05$ — ймовірність того, що програма не розпізнала цифру "3".
- $P(B | A') = 0.02$ — ймовірність того, що програма не розпізнала не цифру "3".

### Формули
1. Знайдемо $P(B)$:

$$
P(B) = P(B | A) \cdot P(A) + P(B | A') \cdot P(A')
$$

Підставимо значення:

$$
P(B) = (0.05 \cdot 0.2) + (0.02 \cdot 0.8) = 0.01 + 0.016 = 0.026
$$

2. Використаємо формулу Баєса:

$$
P(A | B) = \frac{P(B | A) \cdot P(A)}{P(B)}
$$

Підставимо значення:

$$
P(A | B) = \frac{0.05 \cdot 0.2}{0.026} \approx 0.3846
$$

### Висновок
Ймовірність того, що насправді була подана цифра "3", якщо програма її не розпізнала, становить приблизно **38.46%**.
